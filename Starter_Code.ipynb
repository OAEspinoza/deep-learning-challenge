## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf
```

## Step 1: Preprocess the Data

```python
# 1. Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("Resources/charity_data.csv")
application_df.head()
```

### Variables that are targets for the model: IS_SUCCESSFUL
### Variables that are features for the model: ALL OTHERS

```python
# 2. Drop the non-beneficial ID column 'EIN'
application_df=application_df.drop(['EIN','NAME'],axis=1)
application_df.head()
```

```python
# Print column names
for col in application_df.columns:
    print(col)
```

```python
# 3. Determine the number of unique values in each column.
n=application_df.nunique(axis=0)
print(n)
```

```python
# 4. For columns that have more than 10 unique values, determine the number of data points for each unique value.
```

```python
# Look at APPLICATION_TYPE value counts for binning
counts_for_binning=application_df['APPLICATION_TYPE'].value_counts()
counts_for_binning
```

```python
# 5. Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace=list(counts_for_binning[counts_for_binning<500].index)

# Replace in dataframe
for i in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(i,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
counts_classification=application_df["CLASSIFICATION"].value_counts()
counts_classification
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
counts_classification_1=counts_classification[counts_classification>1]
counts_classification_1
```

```python
# 5. Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications2replace = list (counts_classification[counts_classification<150].index)
classifications2replace

# Replace in dataframe
for i in classifications2replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(i,"Other")
    
# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# 6. Convert categorical data to numeric with `pd.get_dummies`
application_df = pd.get_dummies(application_df,dtype=float)
application_df.head()
```

```python
# 7. Split our preprocessed data into our features and target arrays
y = application_df['IS_SUCCESSFUL'].values
X=application_df.drop('IS_SUCCESSFUL', axis=1).values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state = 42)
```

```python
# 8. Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Step 2: Compile, Train and Evaluate the Model

```python
# 2-6 Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len( X_train_scaled[0])
hidden_nodes_layer1=80
hidden_nodes_layer2=30
hidden_nodes_layer3=1

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation='relu'))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation='relu'))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Check the structure of the model
nn.summary()
print(number_input_features)
```

```python
# 7a. Compile the model
nn.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics=['accuracy'])
```

```python
# 7b. Train the model
fit_model = nn.fit(X_train_scaled,y_train,validation_split=0.15, epochs=100)
```

```python
# 8. Create a callback that saves the model's weights every five epochs.
```

```python
# 9. Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
